```python
from GENE_ML.gene_ml.dataset.ScanData import ScanData2
from config import Config
config = Config()
from GENE_ML.gene_ml.parsers.GENEparser import GENE_scan_parser
parser = GENE_scan_parser(config)
# save_dir = '/home/djdaniel/DEEPlasma/vault/data_gen_tutorial7'
save_dir='/scratch/project_462000451/gene_out/gene_auto_97781/omt_range/'
dataset_x0 = ScanData2(name='jet_97781_ion-scale', config=config, parser=parser, save_dir=save_dir,  split_ratio=[0,0,1], categorise=True, group_var_list=[('_grp_species_0','omn'),('_grp_species_0','omt'),('_grp_species_0','omt'),('_grp_species_1','temp'),('units','Tref'),('units','nref'),('units','Bref')])
```

```python
dataset_x0.df
```

```python
import matplotlib.pyplot as plt
import numpy as np

x0 = dataset_x0.df['x01'].to_numpy().astype('float')
real_density = dataset_x0.df['units-nref'].to_numpy().astype('float')
real_temp = dataset_x0.df['units-Tref'].to_numpy().astype('float')

fig, ax = plt.subplots()
axt = ax.twinx()
ax.plot(x0, real_density,label='ne x10^-19')
ax.set_ylabel('ne x10^-19')
ax.plot(x0, real_temp, label='T keV', color = 'orange')
# ax.set_ylabel('T keV')
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = axt.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2, loc='upper right')
ax.vlines((0.90,0.98), 0, 4.1, color='red')
# axt.vlines((0.90,0.98), 0, 4.1, color='red')
axt.plot(x0, real_density/real_temp, '--', color = 'black', label='ne/Te, (right axis)')
axt.legend()


fig, ax = plt.subplots()
axt = ax.twinx()
ne_grad = np.gradient(real_density, x0)
temp_grad = np.gradient(real_temp, x0)

ax.plot(x0, ne_grad)
ax.plot(x0, temp_grad)
ax.vlines((0.90,0.98), 0, -45, color='red')
axt.plot(x0, ne_grad/temp_grad, '--', color = 'black', label='ne_grad / Te_grad (right axis)')
axt.legend()


```
