```python
import numpy as np
from GENE_ML.gene_ml.samplers.LatinHyperCube import LatinHyperCube
import sympy
max_dim = 8
step = 2
parameters = np.arange(1,max_dim)
parameters_sets = [parameters[0:i] for i in np.arange(3,max_dim,2)]
print(parameters_sets)

n_samp = np.logspace(1,4,11).astype('int')
import matplotlib.pyplot as plt
plt.title('Unit Hypercube Space')
plt.ylabel('Convex Hull Volume')
plt.xlabel('Number of Samples')

vol_norm_all = []
dis_all = []

for params in parameters_sets:
    vol_norm = []
    dis = []
    for n in n_samp:
        bounds = [(-1,1)]*len(params)
        print('DEBUG', bounds, params, n)
        sampler = LatinHyperCube(bounds, n, params)
        vol_norm.append(sampler.hull_volume_norm())
        dis.append(sampler.get_discrepancy())
    vol_norm_all.append(vol_norm)
    dis_all.append(dis)
    plt.plot(n_samp, vol_norm, label=f'{len(params)}D')
plt.legend()
plt.show()
    
```
